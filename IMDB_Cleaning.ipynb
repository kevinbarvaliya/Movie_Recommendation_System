```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv("IMDB_Final.csv")
```

```python
movies.head()
```

```python
#Removing spaces at the beginning and spaces at the end 
movies["Overview"] = movies["Overview"].apply(lambda x : x.strip())
```

```python
movies.head()
```

```python
#checking Null values
movies.isna().sum()
```

```python
#cheking data types
movies.dtypes
```

```python
#sorting values by year
#Here there are some unknown values in the year column
#Because of these values we cant convert year column into numeric, so we have to remove these values.

movies.sort_values(["Year"])
```

```python
#Here,we can see round brackets and roman numeral values instead of year
movies["Year"].sort_values()
```

```python
movies["Year"].sort_values(ascending = False)
```

```python
#Replacing round brackets and roman numeral values with NA

movies["Year"] = movies["Year"].str.replace("(", '')
movies["Year"] = movies["Year"].str.replace("X|V|I", '')
```

```python
movies["Year"].sort_values()
```

```python
movies["Year"].sort_values(ascending = False)
```

```python
#Here. we can see commas between numbers.
#We cant convert these strings into numeric if commas are present between numbers

movies["Time"].sort_values()
```

```python
#Replacing commas with with None
movies["Time"] = movies["Time"].str.replace(',','')
```

```python
#We are rearranging the data fame because we did lots of transformation
movies.sort_index(inplace  = True)
```

```python
movies.head()
```

```python
#there are 35 unknown values in votes column
movies = movies.sort_values(["Votes"]).iloc[35:,].sort_index()
```

```python
#converting columns into numeric 
movies["Year"] = pd.to_numeric(movies["Year"])
movies["Time"] = pd.to_numeric(movies["Time"])
movies["Votes"] = pd.to_numeric(movies["Votes"])
movies["Metascore"] = pd.to_numeric(movies["Metascore"])
movies["Gross(Million_$)"] = pd.to_numeric(movies["Gross(Million_$)"])
```

```python
movies
```

```python
#cheking  duplicated values
movies.duplicated().sum()
```

```python
#Removing duplicated values
movies = movies.drop_duplicates()
```

```python
movies.duplicated().sum()
```

```python
movies.isna().sum()
```

```python
#When I was checking the Director column I realized that some directors' names merged into the actor's column.
#In the director column we have some other text Instead of the directors' name
movies["Director"].value_counts()
```

```python
movies[["Director","Actors"]]
```

```python
movies["Director"].isna().sum()
```

```python
#Replacing unknown text from director column with "actor" 
movies["Director"] = movies["Director"].replace("See full summary", "actor")
movies["Director"] = movies["Director"].replace("Add a Plot", "actor")
movies["Director"] = movies["Director"].replace("See full synopsis", "actor")
```

```python
movies["Director"].value_counts()
```

```python
#Below python code extract director name from actor column and replace "actor" text into a director  column with director name which is extracted from actors column 
movies.loc[movies["Director"] == "actor", "Director"] = movies.loc[movies["Director"] == "actor","Actors"].astype(str).apply(lambda x : x.split(",")[0])
```

```python
movies["Director"].value_counts()
```

```python

movies[movies["Director"] == "nan"]
```

```python
#Replacing nan with NaN
movies["Director"] = movies["Director"].replace("nan",np.nan)
```

```python
# some movies do not have an overview, but still, it has some content 
#we have to replace this content with NAN
movies["Overview"].value_counts()
```

```python
#4 movies have a similar overview.
movies[movies["Overview"] == "Macbeth, the Thane of Glamis, receives a prophecy from a trio of witches that one day he will become King of Scotland. Consumed by ambition and spurred to action by his wife, Macbeth murders his king and takes the throne for himself."]
```

```python
movies["Overview"] = movies["Overview"].replace("Add a Plot","") 
movies["Overview"] = movies["Overview"].replace("Plot unknown.","") 
movies["Overview"] = movies["Overview"].replace("Plot under wraps.","") 
movies["Overview"] = movies["Overview"].replace("Plot kept under wraps.","")  
```

```python
movies["Overview"].value_counts()
```
