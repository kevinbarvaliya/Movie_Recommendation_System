```python
import pandas as pd
import requests
from bs4 import BeautifulSoup 
from flask import Response

```

```python
#pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
#list variables to store scrapped data

Movie_name = []
Genre = []
Overview = []
Year = []
Time = []
Rating = []
Votes = []
Metascore = []
Certificate = []
Gross = []
Director = []
Actors = []
Short_desc = []
```

```python
next_page = "/search/title/?title_type=feature&user_rating=5.0,&num_votes=5000,"
n=1
```

```python
while next_page:
    url = "https://www.imdb.com" + next_page
    respose = requests.get(url) #Making a request to a web page
    soup = BeautifulSoup(respose.content, "html.parser") #pulling HTML code from IMDB web-page
    movie_data = soup.findAll("div", attrs = {'class':'lister-item mode-advanced'}) #sercing 

    for row in movie_data: #This loop will fatch movie data one by one and append it into list.

        print(str(n)+ "  " + row.find("span", class_ = "lister-item-index unbold text-primary").text)

        Movie_name.append(row.h3.a.text)

        Genre.append(row.p.find('span', class_ = 'genre').text[1:] if row.p.find('span', class_ = 'genre') else "" )

        Overview.append(row.findAll("p", attrs = {"class" : "text-muted"})[1].text)


        Year.append(row.h3.find('span', class_ = 'lister-item-year text-muted unbold').text[-5:-1])


        Time.append(row.p.find('span', class_ = 'runtime').text[:-4] if row.p.find('span', class_ = 'runtime') is not None else "" )


        Rating.append(row.strong.text if row.strong is not None else "" )


        Certificate.append(row.p.find("span", class_= "certificate").text if row.p.find("span", class_= "certificate") else "") 


        if row.findAll("span", attrs = {'name':'nv'}):
            val = row.findAll("span", attrs = {'name':'nv'})
            Votes.append(val[0].text.replace(",",""))
            Gross.append(val[1].text[1:-1] if len(val)>1 else "")   
        else:
            Votes.append("")
            Gross.append("")


        Metascore.append(row.find('span', class_ = 'metascore').text.strip() if row.find('span', class_ = 'metascore') else "" )

        i = 13
        try:
            Director.append(row.findAll('a', attrs={'class':''})[i].text)
        except IndexError:
            Director.append("")


        actors_len = len(row.findAll('a', attrs={'class':''}))
        i+=1
        actor_list = []
            while i<actors_len:
                if row.findAll('a', attrs={'class':''})[i] is not None:
                    actor_list.append(row.findAll('a', attrs={'class':''})[i].text)
            else:
                break
            i += 1
        Actors.append( ",".join(actor_list))

        next_page = soup.findAll('a', attrs={'class':'lister-page-next next-page'})[0]['href'] if soup.findAll('a', attrs={'class':'lister-page-next next-page'}) else None
        n += 1
        
```

```python
#creating DataFrame from all list variables in which we stored movie information.
df = pd.DataFrame({"Movie_name" : Movie_name,
              "Genre" : Genre,
              "Overview" : Overview,
              "Year" : Year,
              "Time" : Time,
              "Rating" : Rating,
              "Certificate" : Certificate,
              "Votes" : Votes,
              "Gross(Million_$)" : Gross,
              "Metascore" : Metascore,
              "Director" : Director,
              "Actors" : Actors})
```
