```python
import pandas as pd
```

```python
movies = pd.read_csv('IMDb_Clean.csv')
```

```python
movies.head()
```

```python
movies["Genre"]
```

```python
movies = movies.sort_values(["Votes"], ascending=False).head(10000)
```

```python
movies = movies[["Movie_name", "Genre", "Overview", "Director", "Actors"]]
```

```python
movies["Overview"] = movies["Overview"].str.replace('[^\dA-Za-z0-9 ]', '')
```

```python
movies.head()
movies.isna().sum()
```

```python
movies["Genre"] = movies["Genre"].str.replace(","," ")
```

```python
movies.head()
movies.isna().sum()
```

```python
movies["Director"] = movies["Director"].str.replace(" ","")
```

```python
movies
```

```python
movies['Actors'] = movies['Actors'].str.replace(" ", "")
movies['Actors'] = movies['Actors'].str.replace(",", " ")
movies
```

```python
movies["tag"] = movies["Overview"].map(str) + " " + movies["Actors"].map(str)
```

```python
movies.reset_index(inplace=True)
```

```python
movies.head()
movies.isna().sum()
```

```python
movies.isna().sum()
```

```python
movies["tag"]
```

```python

new = movies.drop(columns=['Overview','Actors','index'])
new
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(max_features=1000,stop_words='english')

tfidf = vectorizer.fit_transform(new['tag'])
```

```python
words =  pd.DataFrame(vectorizer.vocabulary_.items()).sort_values(1)
words
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=1000,stop_words='english')
```

```python
new['tag'].isna().sum()
```

```python
vector = cv.fit_transform(new['tag']).toarray()
```

```python
movies['Genre'] = movies['Genre'].apply(lambda x : x.strip())
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
onehot_encoder = OneHotEncoder(sparse=False)
onehot = onehot_encoder.fit_transform(movies[['Director','Genre']])
print(onehot)
```

```python
onehot_encoder.categories_
```

```python
onehot.shape
```

```python
vector.shape
```

```python
import numpy as np
merged_vectore = np.hstack((vector,onehot))
```

```python
merged_vectore.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(merged_vectore)
```

```python
similarity
```

```python
new[new['Movie_name'] == 'The Dark Knight'].index[0]
```

```python
def recommend(movie):
    index = new[new['Movie_name'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].Movie_name)
```

```python
recommend('Avatar')
```
